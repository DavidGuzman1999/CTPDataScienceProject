```python
import geopy
import pandas as pd
```

```python
df = pd.read_csv("Data/NYPD_Complaint_Data_Current__Year_To_Date_.csv")
df.head()
```

```python
df.shape
```

```python
# creating a new dataframe with just lat and long 
df_latLong = df[['Latitude', "Longitude"]].copy()
df_latLong.head()
```

```python
print (df_latLong.shape)
```

```python
geolocator = geopy.Nominatim(user_agent="check_1")
```

```python
def get_zip_code(x):
    location = geolocator.reverse("{}, {}".format(x["Latitude"],x["Longitude"]))
    return location.raw["address"]["postcode"]
df_latLong["zipcode"] = df_latLong[0:100].apply(lambda x: get_zip_code(x), axis=1)
df_latLong.head()
    
```
