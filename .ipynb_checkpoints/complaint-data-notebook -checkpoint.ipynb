### The Borough results should be presented by rate of complaints per population size

```python
# How do you calculate rate per population?
# A complaint rate is calculated by dividing the number
# of reported complaints by the total population. The result
# is then multiplied by 100,000. For example, in the course of a year there
# were 58,372 complaint in Brooklyn and the population was
# 2,736,074. This equals a complaint rate of 2133.4 per 100,000.
```

```python
bk_result = (58372/2736074)*100000
bk_result
```

```python
# Borough population by United States 2020 census
# Queens pop 2405464
# Brooklyn pop 2,736,074 
# Manhattan pop 1,694,251
# Brox pop 1472654
# Staten Island pop 495747
boro_pop_list = [2736074, 1694251, 2405464, 1472654, 495747]
print(boro_pop_list) # borough population list
print(boro_key_list) # borough name list
print(boro_value_list)# total number of complaints
```

### Improve this step by using the new dataframe (df_cmp_boro) as lists

```python
# Making borough dictionary with name=>complaint_value as key pair
zip_iterator = zip(boro_key_list, boro_value_list)
boro_dict = dict(zip_iterator)
print(boro_dict)
```

```python
# Making boro dicitonary with name=>total population as key pair
zip_iterator = zip(boro_key_list, boro_pop_list)
boro_pop_dict = dict(zip_iterator)
print(boro_pop_dict)

# for reference
# how to sort a dict by value
#{key: value for key, value in sorted(boro_pop_dict.items(), key = lambda item: item[1],reverse = True)}
```

```python
# Create a function that compares the keys (boro_name) of two dictionaries.
# get the rate list after dividing by population and muliplying by 100,000
# function uses as input two dictionaries(one with total population and one with total number of complaints)
def cmpl_per_100k(dict_cmpl_num, dict_pop_num):
    cmpl_div_list =[]
    for key in dict_cmpl_num: # traverse first dict
        if(key in dict_pop_num and key == key): # compare key from dict1 and dict2
            cmpl_div_list.append(dict_cmpl_num[key]/dict_pop_num[key]) # Divide complaint count by populaion
    cmpl_per_pop = []
    for item in cmpl_div_list:
        cmpl_per_pop.append(int(item * 100000))
    
    return cmpl_per_pop
# function returns the complaint count by borough per 100,000
```

```python
cmpl_per = cmpl_per_100k(boro_dict,boro_pop_dict)
```

```python
df_cmp_boro = pd.DataFrame({"boro_name":boro_key_list,"complaint_value":cmpl_per})
df_cmp_boro.sort_values(by = "complaint_value",ascending = False, inplace = True)
```

```python
# lists to be used as input for function label_values()
boro_list2 = list(df_cmp_boro["boro_name"])
value_list2 = list(df_cmp_boro["complaint_value"])
print(boro_list2)
print(value_list2)
```
